```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sympy as sym
sym.init_printing(use_unicode=True)
import os
import os.path as path
import wget
```

```python
def Lagrange(x,X,i):
    
    L = 1
    
    for j in range(X.shape[0]):
        if i != j:
            L *= (x - X[j])/(X[i]-X[j])      
    return L

def Interpolate(x,X,Y):
    
    Poly = 0
    
    for i in range(X.shape[0]):
        Poly += Lagrange(x,X,i)*Y[i]
        
    return Poly

```
