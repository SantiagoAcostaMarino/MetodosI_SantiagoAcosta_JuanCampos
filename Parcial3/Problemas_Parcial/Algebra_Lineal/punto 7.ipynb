```python
import numpy as np
```

```python
X=np.array([150,200,300])
A=np.array([[3,-1,-1],[-1,3,1],[2,1,4]])
b=[1,3,7]
w=1.01
```

```python
def sor_iterative(A,X,b,w,e=1*(10**-6)):
    X_0=X
    
    n=0

    while n<100:
        for i in range(3):
            if i==0:
                X_0[i]=(1-w)*(X)[i]+(w/A[i][i])*(b[i]-A[i][i+1]*X[i+2]-A[i][i+2]*X[i+2])
                
            elif i==1:
                X_0[i]=(1-w)*(X)[i]+(w/A[i][i])*(b[i]-A[i][i-1]*X[i-1]-A[i][i+1]*X[i+1])
            else:
                X_0[i]=(1-w)*(X)[i]+(w/A[i][i])*(b[i]-A[i][i-1]*X[i-1]-A[i][i-2]*X[i-2])
        X=X_0
        n=n+1    
        
    
    return (X_0,n)
```

```python
sor_iterative(A,X,b,w)
```
