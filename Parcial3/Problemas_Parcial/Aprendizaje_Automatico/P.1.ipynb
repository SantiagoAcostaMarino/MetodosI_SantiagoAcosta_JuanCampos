```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Variables y función
x, y = sym.symbols('x y')
f = x**2 - y**2 + 2*x

# Punto en el que queremos encontrar el plano tangente
a = (1, 1)

# Calcular el valor de la función en el punto a
fa = f.subs({x: a[0], y: a[1]})

# Calcular el gradiente de la función
grad_f = [f.diff(var) for var in (x, y)]
grad_f_at_a = [g.subs({x: a[0], y: a[1]}) for g in grad_f]

# Polinomio de Taylor de primer orden
taylor = fa + sum([grad_f_at_a[i] * (var - a[i]) for i, var in enumerate((x, y))])

# Valor deseado de z para el plano tangente
z_value = 3

# Ecuaración del plano tangente
plane_equation = taylor - z_value

# Imprimir la ecuación del plano tangente
print("El plano tangente es:", sym.Eq(plane_equation, 0))
```

```python
# Definir la función f(x, y) y el plano tangente
def func(x, y):
    return x**2 - y**2 + 2*x

def plano_tangente(x, y):
    return 4 * (x - 1) - 2 * (y - 1) + 2

# Crear un rango de valores para x y y
x = np.linspace(-2, 4, 400)
y = np.linspace(-2, 4, 400)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función y el plano tangente en cada punto
Z_original = func(X, Y)
Z_tangente = plano_tangente(X, Y)

# Crear una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie original
sup_original = ax.plot_surface(X, Y, Z_original, cmap='viridis', alpha=0.7)

# Dibujar el plano tangente
plano_tangent_dibujo = ax.plot_surface(X, Y, Z_tangente, color='r', alpha=0.5)

plt.show()
```
