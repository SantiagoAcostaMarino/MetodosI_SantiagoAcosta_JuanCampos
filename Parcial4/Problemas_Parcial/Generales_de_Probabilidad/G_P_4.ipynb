```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def prob_cumpleaños(n):
    # Calcular la probabilidad de que n personas tengan cumpleaños diferentes
    prob = 1
    for i in range(n):
        prob *= (365 - i) / 365.0
    return 1 - prob

# Calcular la probabilidad para n en el rango de 1 a 80
n_values = np.arange(1, 81)
probabilities = np.array([prob_cumpleaños(n) for n in n_values])


# Grafica
plt.plot(n_values, probabilities, marker='o')
plt.title('Probabilidad de coincidencia de cumpleaños')
plt.xlabel('Número de personas (n)')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()
```
