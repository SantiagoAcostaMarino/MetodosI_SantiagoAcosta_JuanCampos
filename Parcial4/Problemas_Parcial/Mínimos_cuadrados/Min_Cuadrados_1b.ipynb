```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linea1(x):
    return 2*x - 2

def linea2(x):
    return 0.5 - 0.5*x

def linea3(x):
    return 4 - x

# Generar datos para la búsqueda iterativa
x_vals = np.arange(-5, 5, 0.01)
y_vals = np.arange(-5, 5, 0.01)

# Inicializar variables para almacenar los resultados de la búsqueda iterativa
min_distancia = float('inf')
coordenadas_min = None
```

```python
# Realizar la búsqueda iterativa
for x in x_vals:
    for y in y_vals:
        distancia = np.linalg.norm([linea1(x), linea2(x), linea3(x)] - np.array([x, y, 1]))
        if distancia < min_distancia:
            min_distancia = distancia
            coordenadas_min = [x, y]

# Mostrar resultados de la búsqueda iterativa
print("Resultado de la búsqueda iterativa:")
print("Coordenadas:", coordenadas_min)
print("Distancia mínima:", min_distancia)
```

```python
# Resolver el sistema de ecuaciones por mínimos cuadrados
A = np.array([[2, -1], [1, 2], [1, 1]])
b = np.array([2, 1, 4])
xsol = np.linalg.lstsq(A, b, rcond=None)[0]

# Mostrar resultado de mínimos cuadrados
print("\nResultado de mínimos cuadrados:")
print("Coordenadas:", xsol)


```

```python
# Graficar las líneas y el punto de intersección encontrado por mínimos cuadrados
plt.figure(figsize=(8, 6))

for l in [linea1, linea2, linea3]:
    plt.plot(x_vals, l(x_vals), ls='--', lw=2)

plt.scatter(xsol[0], xsol[1], color='r', label='Mínimos Cuadrados')
plt.scatter(coordenadas_min[0], coordenadas_min[1], color='b', label='Búsqueda Iterativa')

plt.title('Líneas en R^2 y Puntos de Intersección')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
